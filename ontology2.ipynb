### Ajout d'une rÃ¨gle qui dÃ©termine l'appartenance d'un Object_Geo a une zone.

```python
onto = get_ontology("onto_herelles.owl").load()
with onto :    
    #je crÃ©Ã© une regle qui oblige un object_geo d'avoir un seul State_geo:    
    Image.equivalent_to.append(Image & (composed_by.only(Elements) & composed_by.some(Elements)) & (has_rules.only(Zone) & has_rules.some(Zone)) )
    Elements.equivalent_to.append(Elements & (has_state.only(State_Geo) & has_state.min(1,State_Geo) & has_state.max(1,State_Geo)))
    
    Object_Geo.equivalent_to.append(Object_Geo & (has_pixels.only(Pixel) & has_pixels.some(Pixel)))
    Object_Geo.equivalent_to.append(is_in.only(Zone) & is_in.some(Zone))

    Pixel.equivalent_to.append(Pixel & (is_pixel_of.only(Object_Geo) & is_pixel_of.min(1,Object_Geo) & is_pixel_of.max(1,Object_Geo)))
    Pixel.equivalent_to.append(is_pixel_of_zone.only(Zone) & is_pixel_of_zone.min(1,Zone) & is_pixel_of_zone.max(1,Zone))
    Pixel.equivalent_to.append(is_pixel_of_zone.only(Zone) & is_pixel_of_zone.min(1,Zone) & is_pixel_of_zone.max(1,Zone))
    Zone.equivalent_to.append(Zone & contains_object.only(Object_Geo))
    Zone.equivalent_to.append(z_has_pixels.only(Pixel) & z_has_pixels.some(Pixel))
    ZoneEau.equivalent_to.append(Zone & destine.only(Eau))
    ZoneEau.equivalent_to.append(destine.min(1,Eau))
    ZoneArtificialisee.equivalent_to.append(Zone & destine.only(Surface_artificialisee))
    ZoneArtificialisee.equivalent_to.append(destine.min(1,Surface_artificialisee))

    ###################
    ###Dans Zone :
    ###contains_object.some(Object_Geo))
    ###c'est pour pas oubliÃ©, j'hesite entre le fait qu'une zone contient obligatoirement au moins une zone ou non
    ###################

    State_Geo.equivalent_to.append(State_Geo & (has_time.only(Valid_Time) & has_time.min(1,Valid_Time) & has_time.max(1,Valid_Time)))
    
    #########################################################
    ### a rajouter plus tard : valid time
    ### Valid_Time.equivalent_to.append(Valid_Time & (...))
    #########################################################

    #Et maintenant je fais en sorte que les sous-concepts de State_Geo soient disjoint les uns des autres :
    AllDifferent([Eau, Espaces_agricoles, Espaces_forestiers_et_naturels, Surface_artificialisee])  

    #je crÃ©Ã© une regle SWRL permettant de connaitre l'appartenance d'un objet a une zone:
    swrl_rule = """
    Image(?i),
    has_pixels(?o, ?p) , z_has_pixels(?z, ?p) -> is_in(?o, ?z)"""

    rule = Imp("DetectionObjectIsInZone")
    rule.set_as_rule(swrl_rule)

    #sync_reasoner(infer_property_values = True, debug=2)
    
    onto.save(file = "onto_herelles.owl", format = "rdfxml")
```
